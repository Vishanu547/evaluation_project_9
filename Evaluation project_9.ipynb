```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/DSData/master/Telecom_customer_churn.csv')
```

```python
data = data.drop(columns=['customerID'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc = LabelEncoder()
```

```python
data1 = lab_enc.fit_transform(data['gender'])
data2 = lab_enc.fit_transform(data['Partner'])
data3 = lab_enc.fit_transform(data['Dependents'])
data4 = lab_enc.fit_transform(data['PhoneService'])
data5 = lab_enc.fit_transform(data['MultipleLines'])
data6 = lab_enc.fit_transform(data['InternetService'])
data7 = lab_enc.fit_transform(data['OnlineSecurity'])
data8 = lab_enc.fit_transform(data['OnlineBackup'])
data9 = lab_enc.fit_transform(data['DeviceProtection'])
data10 = lab_enc.fit_transform(data['TechSupport'])
data11 = lab_enc.fit_transform(data['StreamingTV'])
data12 = lab_enc.fit_transform(data['StreamingMovies'])
data13 = lab_enc.fit_transform(data['Contract'])
data14 = lab_enc.fit_transform(data['PaperlessBilling'])
data15 = lab_enc.fit_transform(data['PaymentMethod'])
data16 = lab_enc.fit_transform(data['TotalCharges'])
data17 = lab_enc.fit_transform(data['Churn'])
```

```python
data['gender'] = data1
data['Partner'] = data2
data['Dependents'] = data3
data['PhoneService'] = data4
data['MultipleLines'] = data5
data['InternetService'] = data6
data['OnlineSecurity'] = data7
data['OnlineBackup'] = data8
data['DeviceProtection'] = data9
data['TechSupport'] = data10
data['StreamingTV'] = data11
data['StreamingMovies'] = data12
data['Contract'] = data13
data['PaperlessBilling'] = data14
data['PaymentMethod'] = data15
data['TotalCharges'] = data16
data['Churn'] = data17
```

```python
plt.figure(figsize = [20,15],facecolor='g')
plotno = 1


for column in data:
    if plotno<=20:
        ax = plt.subplot(4,5,plotno)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=12)
        
    plotno+=1
    
    
plt.show()    
```

```python
x = data.drop(columns=['Churn'])
y = data['Churn']
```

```python
plt.figure(figsize = (25,35))
plno = 1

for column in x:
    if plno<=20:
        plt.subplot(4,5,plno)
        sns.stripplot(y,x[column])
        
        
    plno+=1    

plt.show()   
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
scalar = StandardScaler()
x_scaled = scalar.fit_transform(x)
```

```python
x_scaled.shape[1]
```

```python
vif = pd.DataFrame()
vif['vif'] = [variance_inflation_factor(x_scaled,i)for i in range(x_scaled.shape[1])]
vif['features'] = x.columns
vif
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size = 0.25,random_state = 19)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
y_pred = log_reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
accuracy
```
